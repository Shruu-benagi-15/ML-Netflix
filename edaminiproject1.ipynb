```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies=pd.read_csv('netflix_titles.csv')
movies.head()
```

```python
# 1. DISPLAY THE FIRST FEW ROWS
movies.head()
```

```python
# 2. SUMMARY STATISCTICS

movies.describe()
```

```python
# 3. CHECK FOR MISSING VALUES

movies.isnull().sum()
```

```python
# 4. DATA TYPES OF EACH COLUMN

movies.info()
```

```python
#5. EXPLORATORY VISUALIZATIONS - HISTOGRAM

movies.hist(bins=30, figsize=(8,4))
plt.show()
```

```python
# CORRELATION MATRIX

corr_matrix=movies.corr()
plt.figure(figsize=(8,4))
sns.heatmap(corr_matrix, annot=True, cmap='Coolwarm', fmt='.2f')
plt.title('Correlaion Matrix')
plt.show()
```

```python
# SCATTER PLOT

sns.boxplot(x='Categorical_Column', y='Numerical_coloumn', data=movies)
plt.title('Box Plot')
plt.show()
```
